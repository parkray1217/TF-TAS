## Data normalization and repeat the experiments

```python
def maxminNormalization(x):
    x = (x-np.min(x))/(np.max(x)-np.min(x))
    return x
```

```python
# change all the data into 0-1 (min-max) 
def data_load_dss(data1,i):
    x = data1["sum_embed"][:i]
    x1 = data1["sum_num_heads"][:i]
    x2 = data1["sum_mlp_ratio"][:i]
    x_dss=data1['dss score'][:i]
    flops =data1['FLOPS'][:i]
    params = data1['Params'][:i]
#     x_retrain_50=data1['retrain accuracy50'][:i]
    x_retrain_100=data1['retrain accuracy100'][:i]
    y = np.array(x)
    y1 = np.array(x1)
    y2 = np.array(x2)
    y3 = np.array(x_dss)
    y4 = np.array(flops)
#     y4 = np.array(x_retrain_50)
    y5 = np.array(x_retrain_100)
    y6 = np.array(params)
    minmax_data_head=maxminNormalization(y1)
    minmax_data_embed=maxminNormalization(y)
    minmax_data_mlp=maxminNormalization(y2)
    Normed_dss_score = maxminNormalization(y3)
#     Normed_retrain_50_score = maxminNormalization(y4)
    Normed_flops = maxminNormalization(y4)
    Normed_retrain_100_score = maxminNormalization(y5)
    Normed_params = maxminNormalization(y6)
#     return minmax_data_head,minmax_data_embed,minmax_data_mlp,Normed_dss_score,x_dss,Normed_retrain_50_score,Normed_retrain_100_score
    return minmax_data_head,minmax_data_embed,minmax_data_mlp,Normed_dss_score,x_dss,Normed_flops,Normed_retrain_100_score,Normed_params
```

```python
def data_load_snip(data1,i):
    x = data1["sum_embed"][:i]
    x1 = data1["sum_num_heads"][:i]
    x2 = data1["sum_mlp_ratio"][:i]
    x_snip=data1['snip score'][:i]
    y = np.array(x)
    y1 = np.array(x1)
    y2 = np.array(x2)
    y3 = np.array(x_snip)
    minmax_data_head=maxminNormalization(y1)
    minmax_data_embed=maxminNormalization(y)
    minmax_data_mlp=maxminNormalization(y2)
    Normed_snip_score = maxminNormalization(y3)
    return minmax_data_head,minmax_data_embed,minmax_data_mlp,Normed_snip_score,x_snip
```

```python
def data_load_grasp(data1,i):
    x = data1["sum_embed"][:i]
    x1 = data1["sum_num_heads"][:i]
    x2 = data1["sum_mlp_ratio"][:i]
    x_grasp=data1['grasp score'][:i]
    y = np.array(x)
    y1 = np.array(x1)
    y2 = np.array(x2)
    y3 = np.array(x_grasp)
    minmax_data_head=maxminNormalization(y1)
    minmax_data_embed=maxminNormalization(y)
    minmax_data_mlp=maxminNormalization(y2)
    Normed_grasp_score = maxminNormalization(y3)
    return minmax_data_head,minmax_data_embed,minmax_data_mlp,Normed_grasp_score,x_grasp
```

```python
# Spearman score & p value
def calculate_spearman_score(array1,array2):

    rho, pval = stats.spearmanr(array1, array2)

    return rho
```

```python
# p value
def calculate_spearman_p(array1,array2):

    rho, pval = stats.spearmanr(array1, array2)

#     print("Spearman correlation coefficient:", rho)
#     print("p-value:", pval)    
    return pval
```

```python
# plot the correlation scatter plot
import numpy as np
import matplotlib.pyplot as plt
def dss_plot_plus(x_dss,normed_data_embed,normed_data_mlp,normed_data_head):

    fig, axs = plt.subplots(2, 3, figsize=(30,18 ))
    
    # added embed and mlp/heads
    normed_data_head_embed = normed_data_head+normed_data_embed
    normed_data_mlp_embed = normed_data_mlp+normed_data_embed
    normed_data_mlp_head = normed_data_mlp+normed_data_head
#     rho,pval = calculate_spearman(array1,array2)
    
    # x: dss score y:nomarlized sum of heads
    axs[0,0].scatter(x_dss, normed_data_head)
    axs[0,0].set_title('MHSA_NORM\nspearman correlation score='+str(calculate_spearman_score(normed_data_head,x_dss))+"\np value="+str(calculate_spearman_p(normed_data_head,x_dss)))
    axs[0,0].set_xlabel('dss score')
    axs[0,0].set_ylabel('heads')
    axs[0,0].grid(True, linestyle='--', color='gray')

    
    # x: dss score y:nomarlized sum of mlp
    axs[0,1].scatter(x_dss, normed_data_mlp)
    axs[0,1].set_title('MLP_NORM\nspearman correlation score='+str(calculate_spearman_score(normed_data_mlp,x_dss))+"\np value="+str(calculate_spearman_p(normed_data_mlp,x_dss)))
    axs[0,1].set_xlabel('dss score')
    axs[0,1].set_ylabel('mlp')
    axs[0,1].grid(True, linestyle='--', color='gray')
    
    
    # x: dss score y:nomarlized sum of embed
    axs[0,2].scatter(x_dss, normed_data_embed)
    axs[0,2].set_title('EMBED_NORM\nspearman correlation score='+str(calculate_spearman_score(normed_data_embed,x_dss))+"\np value="+str(calculate_spearman_p(normed_data_embed,x_dss)))
    axs[0,2].set_xlabel('dss score')
    axs[0,2].set_ylabel('embed')
    axs[0,2].grid(True, linestyle='--', color='gray')
    
    # x: dss score y:nomarlized sum of embed+nomarlized sum of heads
    axs[1,0].scatter(x_dss, normed_data_head_embed)
    axs[1,0].set_title('MHSA_EMBED_NORM\nspearman correlation score='+str(calculate_spearman_score(normed_data_head_embed,x_dss))+"\np value="+str(calculate_spearman_p(normed_data_head_embed,x_dss)))
    axs[1,0].set_xlabel('dss score')
    axs[1,0].set_ylabel('heads+embed')
    axs[1,0].grid(True, linestyle='--', color='gray')

    # x: dss score y:nomarlized sum of embed+nomarlized sum of mlp
    axs[1,1].scatter(x_dss, normed_data_mlp_embed)
    axs[1,1].set_title('MLP_EMBED_NORM\nspearman correlation score='+str(calculate_spearman_score(normed_data_mlp_embed,x_dss))+"\np value="+str(calculate_spearman_p(normed_data_mlp_embed,x_dss)))
    axs[1,1].set_xlabel('dss score')
    axs[1,1].set_ylabel('mlp ratio+embed')
    axs[1,1].grid(True, linestyle='--', color='gray')
    
    # x: dss score y:nomarlized sum of heads+nomarlized sum of mlp
    axs[1,2].scatter(x_dss, normed_data_mlp_head)
    axs[1,2].set_title('MHSA_MLP_NORM\nspearman correlation score='+str(calculate_spearman_score(normed_data_mlp_head,x_dss))+"\np value="+str(calculate_spearman_p(normed_data_mlp_head,x_dss)))
    axs[1,2].set_xlabel('dss score')
    axs[1,2].set_ylabel('mlp ratio+heads')
    axs[1,2].grid(True, linestyle='--', color='gray')

    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
    plt.show()
```
